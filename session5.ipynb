```python
try:
    age=int(input("Enter your age"))
    if age<0 or age>100:
        raise ValueError("age must be between 0 to 100")
    print("age is verified")
except ValueError as e:
    print(e)
```

# finally 

```python
try:
    age=int(input("Enter your age"))
    if age<0 or age>100:
        raise ValueError("age must be between 0 to 100")
    print("age is proper")
except ValueError as e:
    print(e)
finally:
    print("Finally age is verified")
```
