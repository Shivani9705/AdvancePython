```python
# Write a function in python to calculate hypotenuse of a right 
# angled triangle.
```

```python
import math
```

```python
def cal_hypo(s1,s2):
    c=(s1**2+s2**2)
    sq=math.sqrt(c)
    print(f"Hypotaneus of right angle triangle is {sq}")
```

```python
cal_hypo(4,3)
```

```python
def verify_triangle(a,b,c):
    if(a<(b+c) and b<(a+c) and c<(a+b)):
        print("sides are forming triangle")
    else:
        print("sides are not forming triangle")
```

```python
verify_triangle(12,14,18)
```
