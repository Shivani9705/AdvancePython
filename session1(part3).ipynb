```python
# generator function 
# it is special type of function that returns an iterator 
# it has two types 
# 1)yield 
# 2)next 
```

```python
def gen_function():
    a,b=0,1
    while True:
        yield a
        a,b=b,a+b
```

```python
data=gen_function()
```

```python
for i in range(1,5):
    print(next(data))
```

```python
# map function 
# the map function is a built in function in python that applies a given
# function to each itom in an iterable(list,tuple,string)
```

```python
# map(function,iterables...)
```

```python
def square(x):
    return x*x

lst=[1,2,3,4,5,6,7]
```

```python
sq=map(square,lst)
```

```python
print(list(sq))
```
