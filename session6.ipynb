# Multithreading = multiple threads to handle user 

```python
import threading 
```

```python
def print_message():
    print("Welcome to ETLHIVE")
```

```python
t1=threading.Thread(target=print_message)
```

```python
t1.start()
```

```python
def print_message(name):
    print("Welcome to ETLHIVE",name)
```

```python
t2=threading.Thread(target=print_message,args=("shakshi",))
```

```python
t2.start()
```

```python
lst=[2,3,4,55,6,77,88,9,]
```

```python
def cal_sq(s):
    print(f"the square of number {s} is {s*s}")
```

```python

for i in lst:
    th=threading.Thread(target=cal_sq,args=(i,))
    th.start()
```

```python
import urllib.request
from urllib.request import urlretrieve
```

```python
def download_file(lnk):
    file_name=lnk.split('/')[-1]
    print("file downloading.......")
    loc="C:/Users/admin/Downloads/"+file_name
    urllib.request.urlretrieve(lnk,loc)
    print(f"{file_name} is downloaded")
```

```python
download_file('https://raw.githubusercontent.com/utkarshg1/mlproject_regression/main/artifacts/train.csv')
```

```python
lst=['https://raw.githubusercontent.com/utkarshg1/mlproject_regression/main/artifacts/train.csv',
     'https://raw.githubusercontent.com/utkarshg1/mlproject_regression/main/artifacts/data.csv',
     'https://raw.githubusercontent.com/utkarshg1/mlproject_regression/main/artifacts/test.csv']
```

```python
for i in lst:
    th=threading.Thread(target=download_file,args=(i,))
    th.start()
```
