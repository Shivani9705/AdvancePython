```python
# Write a function in python to calculate hypotenuse of a right angled triangle.
```

```python
def cal_hypotaneous(a,b):
    c=(a**2+b**2)**0.5
    print("hypotaneous of right angle triangle is =",c)
```

```python
cal_hypotaneous(3,4)
```

```python
# ➢Write a function in python which takes 3 sides and determines whether 3 sides form triangle or
# not .
```

```python
def verify_triangle(a,b,c):
    if (a<b+c) and (b<a+c) and (c<a+b):
        print("Sides are forming triangle ")
    else:
        print("sides are not forming triangle ")
```

```python
verify_triangle(1,2,3)
```

```python
verify_triangle(14,15,16)
```
