```python
# create a class 
class Person:
    def info(self):
        self.name="Prashant"
        self.age=23
    def intro(self):
        print("Hello my name is ",self.name,"and my age is ",self.age)
```

```python
# create an object 
p1=Person()
```

```python
p1.info()
```

```python
p1.intro()
```

```python
# class 
class Car:
    def properties(self,name,color,mileage):
        self.name=name 
        self.color=color 
        self.mileage=mileage 
    def display(self):
        print(f"Name={self.name}\nColor={self.color}\nMileage={self.mileage}")
```

```python
# object 
car1=Car()
```

```python
car1.properties("Audi",'white',12.56)
```

```python
car1.display()
```

```python
car2=Car()
```

```python
car2.properties("swift","black",12.5)
```

```python
car2.display()
```

```python
# create a class  bank balance =5000 
# deposit(5000) total amount=10000
# withdraw(3000) total amount=7000 
```

```python
class Bank:
    def __init__(self):
        self.balance=5000
    def deposit(self,da):
        self.da=da 
        self.balance=self.balance+self.da
        print("Balance amount after deposit is ",self.balance)
    def withdraw(self,wa):
        self.wa=wa
        if self.wa<=self.balance:
            self.balance=self.balance-self.wa 
            print("Balance amount after withdraw is ",self.balance)
        else:
            print("insufficient amount in your bank")
```

```python
b1=Bank()
```

```python
b1.deposit(3000)
```

```python
b1.withdraw(2000)
```
