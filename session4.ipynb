```python
class India:
    def Capital(self):
        print("The capital of india is new delhi")
    def Language(self):
        print("The official language of india is Hindi")
```

```python
class Usa:
    def Capital(self):
        print("The capital of usa is washington DC")
    def Language(self):
        print("The official language of USA is english")
```

```python
I=India()
U=Usa()
```

```python
I.Capital()
I.Language()
U.Capital()
U.Language()
```

```python
for i in I,U:
    i.Capital()
    i.Language()
```

```python
class cal_area:
    def __init__(self,a,b=0):
        self.a=a
        self.b=b
        self.aos=a*a
        self.aor=a*b 
    def area_of_square(self):
        print("Area of Square=",self.aos)
    def are_of_reactangle(self):
        print(f"Area of rectangle={self.aor}")
```

```python
a1=cal_area(5)
```

```python
a1.area_of_square()
```

```python
a1.are_of_reactangle()
```

```python
a2=cal_area(8,9)
```

```python
a2.are_of_reactangle()
```

```python
a2.area_of_square()
```

```python
# operator overloading 
# + = __add__
# - = __sub__
# / =__div__
# < = __lt__ 
# > = __gt__ 
# & = __and__ 
# | = __or__
```

```python
class Add:
    def __init__(self,a):
        self.a=a
    def __add__(self,other):
        return self.a+other.a
```

```python
a1=Add(5)
a2=Add(10)
```

```python
a1+a2
```

```python
a3=Add(20)
a4=Add(50)
```

```python
a3+a4
```

```python
class Compare:
    def __init__(self,a):
        self.a=a
    def __gt__(self,other):
        if self.a > other.a:
            return f"My first object is greater than other"
        else:
            return ("My second object is greater")
            
```

```python
c1=Compare(10)
```

```python
c2=Compare(20)
```

```python
c2>c1
```

```python
class Addition:
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def __add__(self,other):
        return f"{self.a+other.a}+i{self.b+other.b}"
```

```python
obj1=Addition(3,4)
```

```python
obj2=Addition(4,6)
```

```python
obj1+obj2
```
